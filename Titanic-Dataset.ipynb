```python
import pandas as pd
df=pd.read_csv('train.csv')
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
survived_count = (df['Survived'] == 1).sum()
died_count = (df['Survived'] == 0).sum()
print("Total persons survived =",survived_count)
print("Total persons died =",died_count)
```

```python
colours=["darkgreen","firebrick"]
data={
    'Survived': survived_count,
    'Died': died_count
}
plt.bar(data.keys(),data.values(),color=colours,width=0.2)
plt.xlabel("Survival status")
plt.ylabel("Total persons")
plt.show()
```

```python
male_count=(df['Sex']=='male').sum()
female_count=(df['Sex']=='female').sum()
print("Total number of males =",male_count)
print("Total number of females =",female_count)
```

```python
colours=["sandybrown","sienna"]
data={'Males':male_count,'Females':female_count}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title("Gender distribution")
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
males_survived=((df['Survived']==1) & (df['Sex']=='male')).sum()
females_survived=((df['Survived']==1) & (df['Sex']=='female')).sum()
print("Total number of males survived =",males_survived)
print("Total number of females survived =",females_survived)
```

```python
males_died = male_count - males_survived
colours=["sandybrown","sienna"]
data={"Males Died":males_died,"Males Survived":males_survived

}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title("Survival status of Males")
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
females_died = female_count - females_survived
colours=["sandybrown","sienna"]
data={"Females Died":females_died,"Females Survived":females_survived

}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title("Survival status of Females")
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
colours=["sandybrown","sienna"]
data={"Females Survived":females_survived,"Males Survived":males_survived
}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title("Females Survived vs Males Survived")
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
colours=["sandybrown","sienna"]
data={"Males Died":males_died ,"Females Died":females_died ,"Males Survived":males_survived  ,"Females Survived":females_survived
}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title("Survival Status by Gender")
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
null_age=df['Age'].isnull().sum()
print(null_age)
```

```python
mean_age_miss = df[(df['Sex'] == 'female') & (df['Name'].str.contains('Miss'))]['Age'].mean()
mean_age_mrs = df[(df['Sex'] == 'female') & (df['Name'].str.contains('Mrs'))]['Age'].mean()
df.loc[(df['Age'].isnull()) & (df['Sex'] == 'female') & (df['Name'].str.contains('Miss')), 'Age'] = mean_age_miss
df.loc[df['Age'].isnull() & (df['Sex']=='female') & (df['Name'].str.contains('Mrs')),'Age']=mean_age_mrs
missing=df['Age'].isnull().sum()
print(missing)
```

```python
Class1=df[(df['Sex']=='male') & (df['Pclass']==1)]['Age'].mean()
Class2=df[(df['Sex']=='male') & (df['Pclass']==2)]['Age'].mean()
Class3=df[(df['Sex']=='male') & (df['Pclass']==3)]['Age'].mean()
df.loc[(df['Age'].isnull()) & (df['Sex']=='male') & (df['Pclass']==1),'Age']=Class1
df.loc[(df['Age'].isnull()) & (df['Sex']=='male') & (df['Pclass']==2),'Age']=Class2
df.loc[(df['Age'].isnull()) & (df['Sex']=='male') & (df['Pclass']==3),'Age']=Class3
missing=df['Age'].isnull().sum()
print(missing)
```

```python
class1_count=(df['Pclass']==1).sum()
class2_count=(df['Pclass']==2).sum()
class3_count=(df['Pclass']==3).sum()
print("Total passengers in Class 1 :",class1_count)
print("Total passengers in Class 2 :",class2_count)
print("Total passengers in Class 2 :",class3_count)
```

```python
class1_survival=((df['Pclass']==1) & (df['Survived']==1)).sum()
class2_survival=((df['Pclass']==2) & (df['Survived']==1)).sum()
class3_survival=((df['Pclass']==3) & (df['Survived']==1)).sum()
print("No. of Survivors from Class 1 :",class1_survival)
print("No. of Survivors from Class 2 :",class2_survival)
print("No. of Survivors from Class 3 :",class3_survival)
```

```python
class1_died=class1_count-class1_survival

colours=["sandybrown","bisque"]
data={'Survived':class1_survival, 'Died':class1_died}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title('Survival Rate of Class 1')
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
class2_died=class2_count-class2_survival

colours=["sandybrown","bisque"]
data={'Survived':class2_survival, 'Died':class2_died}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title('Survival Rate of Class 2')
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```

```python
class3_died=class3_count-class3_survival

colours=["sandybrown","bisque"]
data={'Survived':class3_survival, 'Died':class3_died}
plt.pie(data.values(),labels=data.keys(),shadow=True,colors=colours,autopct='%2.1f%%')
plt.title('Survival Rate of Class 3')
plt.legend(loc="center left",bbox_to_anchor=(1.2,0.7),labels=data.keys())
plt.show()
```
