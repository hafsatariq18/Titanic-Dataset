```python
import pandas as pd
df=pd.read_csv('train.csv')
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
survived_count = (df['Survived'] == 1).sum()
died_count = (df['Survived'] == 0).sum()
print("Total persons survived =",survived_count)
print("Total persons died =",died_count)
```

```python
colours=["darkgreen","firebrick"]
data={
    'Survived': survived_count,
    'Died': died_count
}
plt.bar(data.keys(),data.values(),color=colours,width=0.2)
plt.xlabel("Survival status")
plt.ylabel("Total persons")
plt.show()
```
